```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
temp_db=create_engine('sqlite:///:memory:')
```

```python
df=pd.DataFrame(data=np.random.randint(
    low=0, 
    high=100, 
    size=(4,4)), 
    columns=['a', 'b', 'c', 'd']
)
df
```

```python
df.to_sql(
    name='new_table',
    con=temp_db,
)
```

```python
new_df=pd.read_sql(
    sql='new_table', 
    con=temp_db,
)
new_df
```

```python
pd.read_sql(sql='Select a, c From new_table', con=temp_db)
```
