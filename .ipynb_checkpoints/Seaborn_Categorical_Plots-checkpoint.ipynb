```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv('./Docs/05-Seaborn/StudentsPerformance.csv')
df.head()
```

```python
plt.figure(
    figsize=(10, 4),
    dpi=200
)
sns.boxplot(
    data=df,
    y='reading score',
    x='parental level of education',
    hue='test preparation course'
)
plt.legend(
    bbox_to_anchor=(1.05, 1)
)
plt.show()
```

```python
plt.figure(
    figsize=(10, 4),
    dpi=200
)
sns.violinplot(
    data=df,
    y='reading score',
    x='parental level of education',
    hue='test preparation course'
)
plt.legend(
    bbox_to_anchor=(1.05, 1)
)
plt.show()
```

```python
plt.figure(
    figsize=(12, 8),
    dpi=100
)
sns.violinplot(
    data=df,
    y='reading score',
    x='parental level of education',
    hue='test preparation course',
    split=True
)
plt.legend(
    bbox_to_anchor=(1.35, 1)
)
plt.show()
```

```python
plt.figure(
    figsize=(10, 4),
    dpi=200
)
sns.swarmplot(
    data=df,
    x='math score',
    y='gender',
    size=2,
    hue='test preparation course',
    dodge=True
)
plt.legend(
    bbox_to_anchor=(1.20, 1)
)
plt.show()
```

```python
plt.figure(
    figsize=(10, 4),
    dpi=200
)
sns.boxenplot(
    data=df,
    x='math score',
    hue='gender',
    y='test preparation course',
)
plt.legend(
    bbox_to_anchor=(1.15, 1)
)
plt.show()
```
