```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
```

```python
a=np.linspace(0,10,11)
a
```

```python
b=a**4
b
```

```python
x=np.arange(0,10)
x
```

```python
y=2*x
y
```

```python
# fig is a figure object.
fig=plt.figure()

# Here first two parameters define the position of the lower left corner in the whole canvas.
# This will have no effect in case the canvas is of equal size to axes.
# Whereas the other two parameters define the height-width ratio of the canvas.
# A better example is in the next cell.
axes=fig.add_axes([0,0,1,1])
axes.plot(a,b)

plt.show()
```

```python
fig=plt.figure()

# Large Axes.
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A values')
axes1.set_ylabel('B values')
axes1.set_title('Power of 4')

# Small Axes.
axes2=fig.add_axes([0.2, 0.2, 0.5, 0.5])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(1,50)
axes2.set_xlabel('A values')
axes2.set_ylabel('B values')
axes2.set_title('Zoomed In')

plt.show()
```
