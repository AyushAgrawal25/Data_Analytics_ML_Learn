```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./../Docs/DATA/gene_expression.csv');
df.head()
```

```python
plt.figure(
    figsize=(8, 6)
)
sns.scatterplot(
    data=df,
    x='Gene One',
    y='Gene Two',
    hue='Cancer Present',
    alpha=0.5
);
```

```python
sns.pairplot(
    data=df,
    hue='Cancer Present'
);
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X=df.drop('Cancer Present', axis=1)
y=df['Cancer Present']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler=StandardScaler()
```

```python
scaled_X_train=scaler.fit_transform(X_train)
scaled_X_test=scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier(
    n_neighbors=1
)
```

```python
knn_model.fit(scaled_X_train, y_train)
```

```python
y_pred=knn_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
len(y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
df['Cancer Present'].value_counts()
```

```python
# See we have to decrease the test error.
# Which is 1-accuracy(from classfication report)
# so we have to choose a k value which increases accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_error_rates=[]

for k in range(1, 30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_X_train, y_train)
    
    y_pred_test=knn_model.predict(scaled_X_test)
    test_error=1-accuracy_score(y_test, y_pred_test)
    test_error_rates.append(test_error)
    
# test_error_rates
```

```python
plt.plot(
    range(1, 30),
    test_error_rates
)
plt.ylabel('ERROR RATE')
plt.xlabel('K Neightbors')
```

### Pipeline

```python
# Pipeline ---> GridSearch CV
```

```python
scaler = StandardScaler()
```

```python
knn = KNeighborsClassifier()
```

```python
knn.get_params().keys()
```

```python
# We will create the operations sequence.
opertations = [('scaler', scaler), ('knn', knn)]
opertations
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(
    steps=opertations
)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_values = list(range(1, 20))
```

```python
# We have to take care when we creating pipeline in combination with GridSearchCV.
# param_grid keys should be named in this format.
# knn__n_neighbours
# knn is the model and the n_neighbors is parameter.
param_grid={
    'knn__n_neighbors':k_values,
}
```

```python
full_cv_classifier=GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5, 
    scoring='accuracy'
)
```

```python
full_cv_classifier.fit(X_train, y_train)
```

```python
full_cv_classifier.best_estimator_
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
full_pred=full_cv_classifier.predict(X_test)
```

```python
print(classification_report(y_test, full_pred))
```
