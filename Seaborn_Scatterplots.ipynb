```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('./Docs/05-Seaborn/dm_office_sales.csv')
df.head()
```

```python
# x and y are columns from data frame
# hue is also column, color of dots represent the different value.
# size is also column, size of dots represent the value of that field.
# style is also column, shape of marker.
plt.figure( figsize=(12,8), dpi=200)
sns.scatterplot(
    x='salary',
    y='sales',
    data=df,
    hue='level of education',
    size='training level',
    style='level of education'
)
plt.sve
plt.show()
```
