```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./Docs/05-Seaborn/StudentsPerformance.csv')
df.head()
```

```python
plt.figure(
    figsize=(10, 6),
    dpi=200
)
sns.jointplot(
    data=df,
    x='math score',
    y='reading score',
    hue='gender'
)
plt.show()
```

```python
plt.figure(
    figsize=(10, 6),
    dpi=200
)
sns.pairplot(
    data=df, 
    hue='gender',
    diag_kind='hist'
)
plt.show()
```
